```python
# importrequired libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# define parameters
alpha1 = 1
alpha2 = 2
beta1 = 0.05
beta2 = 0.05
# parameters for time
t0 = 0
T = 200
N = 4000
tau = 0.05
# initial conditions
x0 = 10
y0 = 1
```

```python
# define functions for ODE (predator-prey model)
def f(t,x,y):
    return alpha1*x - alpha2*x*y

def g(t,x,y):
    return -beta1*y + beta2*x*y

# define explicit euler method using number of steps N
def euler(t0, T, N, x0, y0):
    t = np.linspace(0, T, N + 1) # total N+1 time step from 0 to N
    x = np.empty(len(t)) # len(t) = N
    y = np.empty(len(t))
    h = (T - t0) / N
    t[0] = t0
    x[0] = x0
    y[0] = y0
    
    for i in range(N):
        x[i + 1] = x[i] + h * f(t[i], x[i], y[i])
        y[i + 1] = y[i] + h * g(t[i], x[i], y[i])
        
    return t, x, y # return 3 numpy arrays

# define explicit euler method using step size tau
def euler2(t0, T, tau, x0, y0):
    N = int((T - t0) / tau)
    t = np.linspace(0, T, N + 1) # total N+1 time step from 0 to N
    x = np.empty(len(t)) # len(t) = N
    y = np.empty(len(t))
    t[0] = t0
    x[0] = x0
    y[0] = y0
    
    for i in range(N):
        x[i + 1] = x[i] + tau * f(t[i], x[i], y[i])
        y[i + 1] = y[i] + tau * g(t[i], x[i], y[i])
        
    return [t, x, y] # return a list
```

```python
# create a list for different step size tau
list_tau = [0.05, 0.02, 0.01, 0.005, 0.001]

# create list to store solutions(array)
sol = []
# create a loop for numerical solutions of ODE with different step size
for tau in list_tau:
    array = euler2(t0, T, tau, x0, y0)
    sol.append(array)
```

```python
# create a loop to plot all graphs
for array in sol:
    plt.plot(array[0], array[1],"b--", label="prey")
    plt.plot(array[0], array[2],"r--", label="predator")
    plt.xlabel('time')
    plt.ylabel('numbers')
    plt.title("predator-prey model with step size " + str(array[0][1]))
    plt.legend()
    plt.show()
```
